# Rapport d'analyse des logs

Les résultats de ce notebook reposent sur un ensemble de données construit manuellement dans le code afin de simuler des messages reçus via le protocole LoRaChatt. Ces données sont utilisées pour analyser les logs générés par le test du driver SX127x.

Les messages échangés dans LoRaChat suivent une convention simple permettant d’identifier l’émetteur, le destinataire, le contenu du message, ainsi que les paramètres de signal associés RSSI (Received Signal Strength Indicator, ie la puissance du signal) et SNR (Signal-to-Noise Ratio, ie la qualité du signal).

Le format général est le suivant :
```
SRC@DEST:MSG_ID:MESSAGE  [RSSI:VALUE SNR:VALUE]
```
Par exemple:
```
123@*:0:Hello world!  [RSSI:-42 SNR:10]\
456#1:1:Hi there!  [RSSI:-40 SNR:12]\
789@777:2:Hey!  [RSSI:-45 SNR:8]
```
avec 0 comme nombre par défaut pour le champ DEST, indiquant que le message est destiné à tous les utilisateurs (broadcast).

De cette manière, il est notamment possible de vérifier si les messages sont perdus ou non en comparant les MSG_ID (des numéros non consécutifs indiquent des messages perdus), qui peuvent être causés par des interférences, une mauvaise qualité de signal (indiquée par un RSSI faible ou un SNR négatif), ou d'autres facteurs environnementaux comme des collisions ou une portée limitée.
